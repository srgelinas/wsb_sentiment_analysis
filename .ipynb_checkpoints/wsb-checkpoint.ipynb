## Data Cleaning/Text Preprocessing

To prepare the dataset for sentiment analysis, unnecessary columns and rows with either 'deleted' or 'removed' comments were filtered from the dataset. The WSB comments were then cleaned to remove handles, URLS, special characters, and multiple characters using regex expressions.

## Sentiment Analysis

VADER sentiment intensity analyzer was applied to each row of the WSB data to determine the postive, neutral, and negative sentiment from each comment. WSB sentiment on a given trading day was determined by grouping the dataset by timestamp and mean aggregating the sentiment scores. In the time frame: January 28, 2021 to June 6, 2021 positive sentiment in WSB increased, negative sentiment decreased, and neutral sentiment slightly fluctuated. The number of posts on WSB peaked on January 29, 2021 with over 15,000 posts. Positive sentiment also peaked in WSB during mid-February 2021 and mid-April 2021.

## Classifier

A classification model was trained on mean aggregated WSB post data grouped by timestamp to predict whether the S&P500 had a positive or negative percent return on a given trading day. The baseline model had a testing accuracy of 69.2%, which outperformed the baseline accuracy of 65%. To improve the model, additional words that would be consitered "bullish" (positive) and "bearish" (negative) were added to the lexicon for sentiment analysis, the most important features in the model were identified, and cross validation was conducted to find the optimal parameter for logistic regression. In the final model, the classifier's testing accuracy increased to 76.9%, the true negative rate increased from 3.8% to 11.5%, and the false positive rate decreased from 26.9% to 23.1%


## Trading with WSB Sentiment

The percent return from three trading strategies were compared in 2021: using short/long-term simple moving averages of the S&P500 closing price, using short/long-term simple moving averages of the compound WSB sentiment score, and the traditional buy and hold strategy. Buy and hold produced a 13% return, modelling baseline performance. Trading on simple moving averages of the S&P500 closing price and WSB compound sentiment score underperformed buy and hold, with a 10.5% and 7.5% return respectively. WSB sentiment can be used to predict general market trends, as fluctuations in WSB sentiment have some association with changes in market prices. However, trading on fluctuations in WSB sentiment underpforms the market compared to safer, more traditional strategies, such as buy and hold and simple moving averages.

# Data Cleaning / Text Preprocessing

Load the Reddit WSB data, and filter the dataset to the relevant columns: 'title', 'score', 'body', 'comms_num', and 'timestamp'.

```python
df = pd.read_csv('reddit_wsb.csv')
df = df[['title', 'score', 'body', 'comms_num', 'timestamp']]
df.head()
```

Null values in the dataset only appear to be in the 'body' column. When analyzing the 'title' and 'body' columns, rows containing 'removed' and 'deleted' were also treated as null values.

```python
df.info()
df.isnull().sum()
```

Rows containing null values in the 'body' column, as well as rows with deleted or removed title/body text were dropped from the dataset. The 'title' and 'body' columns were then combined into a single column, 'text' after removing the missing rows.

```python
def clean_deleted(text):
    if 'deleted' in text: return ''
    elif 'removed' in text: return ''
    else: return text

#clean deleted title/body
df['title'] = df['title'].apply(clean_deleted)
df['body'] = df['body'].fillna('').apply(clean_deleted)

#combine title and body columns into 'text'
df['text'] = df['title'] + '. ' + df['body']
df['text'] = df['text'].fillna(df['title'])
df.drop(columns = ['title', 'body'], inplace = True)
df.head()
```

The 'text' column was cleaned using regex to remove handles, URLs, special characters, single letters and multiple spaces in the post. Furthermore, a new columm, post length was addeed to the dataset. The 'timestamp' column was cleaned to only contain the date of the post, and dates outside the range of the analysis period were filtered.

```python
def clean_text(text):
    text = text.lower()
    #remove handles
    text = re.sub(r'\\n', '', text)
    #remove URLs
    text = re.sub(r'http\S+', '', text)
    #remove special chars
    text = re.sub(r'[\W]+', ' ', text)
    #remove single letters
    text = re.sub(r'\s+[a-zA-Z]\s+', ' ', text)
    #remove multiple spaces
    text = re.sub(r'\s+', ' ', text)
    return text

#clean the text column
df['cleaned'] = df['text'].apply(clean_text)
#determine post length
df['post_length'] = df['cleaned'].apply(lambda x: len(x.split(' ')))
#extract the date from the timestamp
df['timestamp'] = pd.to_datetime(df['timestamp'])
#remove dates that are out of the scope of analysis
df = df[df['timestamp'].dt.month != 9]
df['timestamp'] = pd.to_datetime(df['timestamp'].dt.date)
df.head()
```

# Sentiment Analysis

WallStreetBet's sentiment may be predictive of overall trends in the stock market. To determine the sentiment from each WSB post, VADER sentiment intensity analyzer was used on each row to determine polarity scores of the 'cleaned' text. The postive, negative, neutral, and compound sentiment scores were extracted for each row and were added to the dataset.

```python
#determine sentiment from cleaned text
SIA = SentimentIntensityAnalyzer()
df['polarity'] = df['cleaned'].apply(SIA.polarity_scores)
df['neg'] = df['polarity'].apply(lambda x: x['neg'])
df['neu'] = df['polarity'].apply(lambda x: x['neu'])
df['pos'] = df['polarity'].apply(lambda x: x['pos'])
df['compound'] = df['polarity'].apply(lambda x: x['compound'])
df = df.drop('polarity', axis = 1)
df.head()
```

Lineplots were plotted to better visualize how the postiive, negative, neutral and compound sentiment scores changed on the WSB subreddit over the time period: January 28, 2021 to June 13, 2021. To remove noise from the graphs, the plots were smoothed to a 5-day moving average of the sentiment score.

```python
plt.figure(figsize=(10,8))

#graph of average compound score over time period: 2021-1-28 to 2021-6-13
plt.subplot(4, 1, 1)
mean_scores = df.groupby('timestamp').mean()['compound']
mean_scores.plot(kind = 'line')
plt.ylabel('Compound Score')
mean_scores.rolling(5).mean().plot(kind = 'line', color = 'red')
ax = plt.gca()
ax.axes.xaxis.set_visible(False)

#positive score
plt.subplot(4, 1, 2)
df.groupby('timestamp').mean()['pos'].plot(kind = 'line')
plt.ylabel('Positive Score')
df.groupby('timestamp').mean()['pos'].rolling(5).mean().plot(kind = 'line', color = 'red')
ax = plt.gca()
ax.axes.xaxis.set_visible(False)

#negative score
plt.subplot(4, 1, 3)
df.groupby('timestamp').mean()['neg'].plot(kind = 'line')
plt.ylabel('Negative Score')
df.groupby('timestamp').mean()['neg'].rolling(5).mean().plot(kind = 'line', color = 'red')
ax = plt.gca()
ax.axes.xaxis.set_visible(False)

#neutral score
plt.subplot(4, 1, 4)
df.groupby('timestamp').mean()['neu'].plot(kind = 'line')
plt.ylabel('Neutral Score')
df.groupby('timestamp').mean()['neu'].rolling(5).mean().plot(kind = 'line', color = 'red')
plt.show()
```

Histograms of the compound, positive, negative, and neutral sentiment scores were graphed along with a fitted density estimate to visualize the distribution of sentiment scores across WSB from January 28, 2021 to June 13, 2021.

```python
#graph of average compound score over time period: 2021-1-28 to 2021-6-13
fig = plt.figure(figsize=(8,4))
sns.distplot(df.groupby('timestamp').mean()['compound'])
sns.distplot(df.groupby('timestamp').mean()['pos'])
sns.distplot(df.groupby('timestamp').mean()['neg'])
sns.distplot(df.groupby('timestamp').mean()['neu'])
fig.legend(labels=['compound','positive', 'negative', 'neutral'])
plt.xlabel('VADER sentiment score')
plt.show()
```

# Baseline Classifier

The baseline classification model will be trained on mean aggregated WSB post data grouped by timestamp to predict whether the S&P500 had a positive or negative percent return on a given trading day.

```python
#split df into target and response variables
X = df_grouped.drop('daily_return', axis = 1)
y = df_grouped['daily_return']
X.head()
```

The logistic regression classifier is trained on 9 numerical features, all standardized for preprocessing. The classifier has aa testing accuracy of 69.2%, which performed above the baseline accuracy of 65.4%: predicting that every trading day, the S&P500 will close with a daily percent gain.

```python
#create pipeline for classifier
num_transformer = Pipeline(steps=[('standardize', StandardScaler())])

preproc = ColumnTransformer(transformers=[\
                        ('numerical', num_transformer, ['neg', 'neu', 'pos', 'compound', 'comms_num', 'score',
                                                       'post_length', 'post_count'])
                        ], remainder = 'drop')

pl = Pipeline(steps=[
                     ('preprocessing', preproc), \
                     ('classifier', LogisticRegression())\
                    ])

#split data into train/test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 27)
#fit pipeline and predict
pl.fit(X_train, y_train)
preds = pl.predict(X_test)
training_acc = np.mean(pl.predict(X_train) == y_train)
testing_acc = np.mean(preds == y_test)
#baseline accuracies to beat
baseline = np.mean(y_test == 0), np.mean(y_test == 1)
print('Training Accuacy: ' + str(training_acc) +
      '\nTesting Accuracy: ' + str(testing_acc) +
      '\nBaseline Accuracies: ' + str(baseline))
```

A confusion matrix was graphed to determine the classifier's true negative, false negative, true positive, and false positive rates. The classifier has a true negative rate of 3.8%, a false negative rate of 0%, a true positive rate of 69.2%, and a false positive rate of 26.9%. 

```python
#plot confusion matrix
cm = confusion_matrix(y_test, preds)
ax= plt.subplot()
sns.heatmap(cm, annot=True, fmt='g', ax=ax)
ax.set_xlabel('Predicted Class')
ax.set_ylabel('True Class')
ax.set_title('Confusion Matrix') 
ax.xaxis.set_ticklabels(['% Loss', '% Gain'])
ax.yaxis.set_ticklabels(['% Loss', '% Gain'])
plt.show()
```

The model's testing accuracy may be a result of a favorable train/test split, so 300 models will be trained to determine the distribution of the classifier's observed testing accuracy. The observed testing accuracy appears to be above the average of simulated testing accuracies.

```python
def get_test_accuracy(pl):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
    pl.fit(X_train, y_train)
    preds = pl.predict(X_test)
    return np.mean(preds == y_test)

clf_accuracy = []
for i in range(300):
    clf_accuracy.append(get_test_accuracy(pl))

plt.hist(clf_accuracy)
plt.scatter(np.mean(preds == y_test), 0, color = 'red')
plt.ylabel('Frequency')
plt.xlabel('Testing Accuracy')
plt.show()
```

The classifier isn't very robust. About 25% of the time, the classifier's testing accuracy outperforms the baseline model's testing accuracy of guessing everyday the S&P500 closes with a percent gain.

```python
def above_baseline():
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
    pl.fit(X_train, y_train)
    preds = pl.predict(X_test)
    acc = np.mean(preds == y_test)
    baseline = np.mean(y_test == 1)
    return acc > baseline

scores = []
for i in range(300):
    scores.append(above_baseline())
np.mean(scores)
```

# Final Model

To improve the classifier, we will add the most common words from WallStreeetBets with specifc bullish (positive) and bearish (negative) sentiment to the lexicon, refit the model, and determine if the testing accuracy increases.

```python
#use count vectorizer to find the most frequent words that aren't stop words
def get_n_counts(df, n):
    document = df.to_numpy()
    vectorizer = CountVectorizer(stop_words = 'english')
    X = vectorizer.fit_transform(document)
    cols = vectorizer.get_feature_names()
    counts = pd.DataFrame(data = X.toarray(), columns = cols)
    return counts.sum(axis = 0).sort_values(ascending = False)[:n].to_dict()

pos_document = df[df['pos'] > df['neg']]['cleaned']
neg_document = df[df['pos'] < df['neg']]['cleaned']
#determine 20 most frequent words among all WSB posts
get_n_counts(df['cleaned'], 20)
```

```python
#determine 20 most frequent words among positive WSB posts
top_20_pos = get_n_counts(pos_document, 20)
top_20_pos
```

```python
#determine 20 most frequent words among negative WSB posts
top_20_neg = get_n_counts(neg_document, 20)
top_20_neg
```

We will avoid adding words to the lexicon that appear in both the 'negative' WSB posts and the 'positive' WSB posts.

```python
#words that are shared betweeen top 20 most common in pos/neg WSB posts   
print([key if key in top_20_neg.keys() else '' for key in top_20_pos.keys()])
```

```python
#bulllish/bearish words added
wsb_dictionary = {
    'call': 1,
    'put': -1,
    'yolo': 1,
    'bag': -1,
    'guh': -1,
    'tendies': 1,
    'launch': 1,
    'pump': 1,
    'dump': -1,
    'diamond': 1,
    'paper': -1,
    'gain': 1,
    'loss': -1,
}

#update sentiment analyzer's lexicon
SIA.lexicon.update(wsb_dictionary)
df['polarity'] = df['cleaned'].apply(SIA.polarity_scores)
df['neg'] = df['polarity'].apply(lambda x: x['neg'])
df['neu'] = df['polarity'].apply(lambda x: x['neu'])
df['pos'] = df['polarity'].apply(lambda x: x['pos'])
df['compound'] = df['polarity'].apply(lambda x: x['compound'])
df = df.drop('polarity', axis = 1)
df_grouped = df.groupby('timestamp').mean()
df_grouped['post_count'] = post_count
```

Feature importances were determined from the baseline classifier. The compound sentiment score and reddit score were the most predictive features of a daily percent gain in the S&P500, while the neutral sentiment score, positive sentiment score, and post length were the most predictive features of a daily percent loss in the S&P500. The number of post comments was the least important feature in our classifier.

```python
#determine feature importance from classifier
importance = pl['classifier'].coef_[0]
features = ['neg', 'neu', 'pos', 'compound', 'comms_num', 'score','post_length', 'post_count']
pd.DataFrame(importance, index = features, columns = ['coefficient'])
```

Despite the classifier having 9 numerical features, removal of the least important feature, number of comments, decreases the model's training and testing accuracy by about 3%. No features were removed from the baseline classifier when constructing the final model.

```python
#split df into target and response variables
X = df_grouped.drop(['daily_return'], axis = 1)
y = df_grouped['daily_return']
```

Refitting the pipeline for the final model, the classifier's training accuracy increased from 68.3% to 70%, and testing accuracy increased from 69.2% to 76.9%.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 27)
pl.fit(X_train, y_train)
preds = pl.predict(X_test)
training_acc = np.mean(pl.predict(X_train) == y_train)
testing_acc = np.mean(preds == y_test)
baseline = np.mean(y_test == 0), np.mean(y_test == 1)
print('Training Accuacy: ' + str(training_acc) +
      '\nTesting Accuracy: ' + str(testing_acc) +
      '\nBaseline Accuracies: ' + str(baseline))
```

An updated confusion matrix was plotted to analyze the performace of the final classifier. Compared to the baseline model, the number of true negative classifications increased from 1 to 3, and the number of false positive classifications decreased from 8 to 6. The final model has a true negative rate of 11.5%, a false negative rate of 0%, a false positive rate of 23.1%, and a true positive rate of 65.4%.

```python
#plot confusion matrix
cm = confusion_matrix(y_test, preds)
ax= plt.subplot()
sns.heatmap(cm, annot=True, fmt='g', ax=ax)
ax.set_xlabel('Predicted Class')
ax.set_ylabel('True Class')
ax.set_title('Confusion Matrix') 
ax.xaxis.set_ticklabels(['% Loss', '% Gain'])
ax.yaxis.set_ticklabels(['% Loss', '% Gain'])
plt.show()
```

To determine the optimal parameter, 'C', for logistic regression, a grid search with 3-fold cross validation was performed. The parameter 'C', had an optimal value of: .12, with a best testing accuracy of 68.3%.

```python
#grid search CV for parameter tuning
parameters = {
    'classifier__C': np.logspace(-5, 5, 50),
}

grid = GridSearchCV(pl, parameters, cv = 3)
grid.fit(X_train, y_train)
grid.best_params_, grid.best_score_
```

The parameter 'C' is robust as shown in the distribution of the histogram below. In 50 grid searches with 3-fold cross validation, the parameter 'C' had an optimal value close to .1, with not much variability. This optimal value is equal to the mode of the histogram of parameter values for 'C'.

```python
params = []
for i in range(50):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
    grid = GridSearchCV(pl, parameters, cv = 3)
    grid.fit(X_train, y_train)
    params.append(grid.best_params_['classifier__C'])
plt.hist(params)
plt.scatter(grid.best_params_['classifier__C'], 0, color = 'red')
plt.show()
```

Finally, a reciever operating characteristic plot was graphed to evaluate the classifier's performace to random guessing. The final classifier has an AUC of .61, which outperforms random guessing's AUC of .5.

```python
# calculate the fpr and tpr for all thresholds of the classification
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 27)
metrics.plot_roc_curve(pl, X_test, y_test)
plt.title('Receiver Operating Characteristic')
plt.plot([0, 1], [0, 1],'r--', label = 'Random Classifier')
plt.plot([0, 0, 1], [0, 1, 1], 'r:', color='black', label='Perfect Classifier')
plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc = 'lower right')
plt.show()
```

# Trading with WSB Sentiment 

With the decreased spending and increased stimulus from the COVID-19 pandemic, the stock market has seen an influx of retail investors like never before. Many of these new retail investors have turned towards Reddit's WallStreetBets, which has seen a drastic surge in subscribers from 1 million to over 10 million in a few months.

We've found evidence supporting WSB's sentiment has predictive value in predicting whether the stock market closes with a daily gain or loss, but can this predictive value from WSB be leveraged to trade stocks? To determine if retail investors should trade off WSB sentiment, the 5 and 20 day moving averages of the compound sentiment score will be plotted alongside the closing price of the S&P500, representing the short term and long term sentiment trends respectively in the subreddit in the form of a teechnical trading indicator.

```python
#visualize simple moving averages of $SPY closing price and WSB sentiment
fig, ax = plt.subplots()

#plot WSB sentiment
ax.plot(df.groupby('timestamp').mean()['compound'].rolling(20).mean(), color='red')
ax.plot(df.groupby('timestamp').mean()['compound'].rolling(5).mean(), color='green')
ax.tick_params(axis='y')
ax.legend(['20 day avg', '5 day avg'], loc = 'upper left')
ax2 = ax.twinx()

#plot SPY closing price and change tick color
ax2.plot(spy_close[1:])
ax2.tick_params(axis='y')
ax2.legend(['Closing Price'], loc = 'upper right')

for ax in fig.axes:
    ax.tick_params(labelrotation=50)

plt.title('WSB Sentiment vs. Price Action')
plt.show()
```

A "buy" signal will be represented by the short term, 5 day moving average, crossing above the long term, 20 day moving average.

A "sell" signal will be represented by the short term, 5 day moving average, crossing below the long term, 20 day moving average.

Trading with WSB sentiment will be compared to trading with the long term and short term simple moving averages of the S&P500's closing price, and a traditional buy and hold strategy. The strategies will then be evaluated by the total account percent return realized during the trading period.

```python
def get_signals(df, long_signal, short_signal):
    """
    A function that determines buy/sell signals from intersections
    of a short term average and long term average signal. A list of
    tuples containing the signal and date index is returned.
    
    """
    diffs = np.diff(np.sign(long_signal - short_signal))
    indicies = np.argwhere(diffs).flatten()
    indicies = indicies[np.isnan(np.sign(long_signal - long_signal)).sum():]
    signals = ['buy' if diffs[index] < 0 else 'sell' for index in indicies]
    return list(zip(signals, indicies))
```

```python
#determine buy/sell signals for $SPY from simple moving averages of $SPY closing price
long_avg = spy_close[1:].rolling(20).mean().to_numpy()
short_avg = spy_close[1:].rolling(5).mean().to_numpy()
sma_signals = get_signals(spy_close, long_avg, short_avg)
sma_signals
```

```python
#determine buy/sell signals for $SPY from simple moving averages of WSB sentiment
long_avg = mean_scores[1:].rolling(20).mean().to_numpy()
short_avg = mean_scores[1:].rolling(5).mean().to_numpy()
wsb_signals = get_signals(mean_scores, long_avg, short_avg)
wsb_signals
```

```python
def buy_shares(shares, capital, price):
    '''
    Buy as many shares (including fractional) with
    available capital
    '''
    shares = capital/price
    capital = 0
    print('buy price: $' + str(price) + '\n'
        'shares purchased: ' + str(shares) + '\n'
        'capital: $' + str(capital) + '\n')
    return shares, capital

def sell_shares(shares, capital, price):
    '''
    Sell all owned shares in portfolio
    '''
    capital = shares*price
    print('sell price: $' + str(price) + '\n'
        'shares sold: ' + str(shares) + '\n'
        'capital: $' + str(capital) + '\n')
    shares = 0
    return shares, capital

def find_next_date(in_table, index, date):
    '''
    Finds the next date in the $SPY table if WSB comments weren't
    made on a trading day.
    '''
    while in_table == 0:
        i = 1
        try:
            date = date + datetime.timedelta(days=i)
            in_table = spy_hist['Date'].isin([date]).sum()
        except ValueError:
            #increment date by 1 if not in table
            i += 1
    return date

def get_price(index, data):
    '''
    Returns the price on a specific day, represented by index.
    '''
    if data == sma_signals:
        price = spy_hist.iloc[index]['Open']
    else:
        #find price on a specific date
        date = mean_scores.index[index]
        in_table = spy_hist['Date'].isin([mean_scores.index[index]]).sum()
        #find next date in table if target isn't present
        if in_table == 0:
            date = find_next_date(in_table, index, date)
        price = spy_hist[spy_hist['Date'] == date]['Open'].item()
    return price

def run_simulation(initial_capital, signals):
    '''
    Simulates a trading period using specific signals and returns
    the final capital and percent return.
    '''
    capital = initial_capital
    shares = 0

    for signal, index in signals:
        #find stock price on a specific date
        price = get_price(index, signals)
        
        if signal == 'buy':
            #buy as many shares with capital
            if capital > 0:
                shares, capital = buy_shares(shares, capital, price)
            else:
                print('not enough capital \n')
        else:
            #sell all owned shares
            if shares > 0:
                shares, capital = sell_shares(shares, capital, price)
            else:
                print('no shares to sell \n')
    
    #sell all shares at the end of trading period
    if capital == 0:
        final_price = spy_hist.iloc[-1]['Open']
        shares, capital = sell_shares(shares, capital, final_price)

    #final return in period
    percent_return = (capital - 10000)/10000 * 100
    print('initial balance: $' + str(10000) + '\n'
        'final balance: ' + str(capital) + '\n'
        'percent return: ' + str(percent_return))
    
    return capital, percent_return
```

```python
#simulate trading with SMA strategy
capital = 10000
run_simulation(capital, sma_signals)
```

```python
#simulate trading with WSB sentiment analysis strategy
spy_hist = spy_hist.reset_index()
capital = 10000
run_simulation(capital, wsb_signals)
```

```python
#simulate trading with buy/hold
capital = 10000
shares = 0

initial_price = spy_hist.iloc[0]['Open']
shares, capital = buy_shares(shares, capital, initial_price)

final_price = spy_hist.iloc[-1]['Open']
shares, capital = sell_shares(shares, capital, final_price)
    
percent_return = (capital - 10000)/10000 * 100
print('initial balance: $' + str(10000) + '\n'
    'final balance: $' + str(capital) + '\n'
    'percent return: ' + str(percent_return))
```
